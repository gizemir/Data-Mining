In this code, frequent sequence mining was performed by applying prefix span to a million data set with a min support value of 1000.

```python
# Installing the required library
!pip install gdown

import gdown
from prefixspan import PrefixSpan

# Download the file from the shared Google Drive link
file_id = "1OULJVNIckohq1Rb9V3sUJ6ByhzahQHzM"
url = f"https://drive.google.com/uc?id={file_id}"
output = "input_file.txt"

# Download file
gdown.download(url, output, quiet=False)

# Reading the file
with open(output, 'r', encoding='utf-8') as file:
    content = file.readlines()

```

```python
# Separating categories and arrays
categories = []
sequences = []
in_sequences = False

for line in content:
    line = line.strip()
    if line.startswith("% Different categories found in input file:"):
        continue
    elif line.startswith("% Sequences:"):
        in_sequences = True
        continue
    elif not in_sequences and line:
        categories.extend(line.split())
    elif in_sequences and line:
        sequences.append([int(x) for x in line.split()])

# Printing results
print(f"Categories ({len(categories)}): {categories}")
print(f"Example User Sequences ({len(sequences)}): {sequences[:5]}")
```

```python
# Create a category map
category_map = {i + 1: category for i, category in enumerate(categories)}
print("Categori map:", category_map)
```

```python
# Sampling to analyze user sequences
print("Example User Sequences:")
for seq in sequences[:5]:
    print([category_map[i] for i in seq])
```

```python
# Find frequent transitions using the PrefixSpan Algorithm
ps = PrefixSpan(sequences)
min_support = 1000  # En düşük destek eşiği (frekans)
frequent_sequences = ps.frequent(min_support)

# Matching frequent transitions with page names
page_names = {i: categories[i-1] for i in range(1, len(categories) + 1)}
converted_sequences = []

for freq, seq in frequent_sequences:
    named_seq = [page_names.get(item, f"Page {item}") for item in seq]
    converted_sequences.append((freq, named_seq))

# Printing the output
print("Frequency | Page Transitions")
print("--------------------------")
for freq, named_seq in converted_sequences:
    print(f"{freq:<8} | {' -> '.join(named_seq)}")
```
